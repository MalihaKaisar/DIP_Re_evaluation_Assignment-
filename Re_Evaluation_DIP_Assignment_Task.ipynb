```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


image1 = cv2.imread('/content/drive/MyDrive/Assignment_DIP/DIP_Lab_task_01.png')
plt.imshow(image1)
plt.show()

print("")

print("Dimension Is: ")
print(np.shape(image1))
```

```python

image_path = "/content/drive/MyDrive/Assignment_DIP/DIP_Lab_task_02.jpg"
image2 = cv2.imread(image_path)

gray_image = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

x, y = 50, 50
sub_region = gray_image[y:y+5, x:x+5]

plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.show()

print("")
print("5x5 Sub-region Intensity Values:")
print(sub_region)
print("")
plt.imshow(sub_region, cmap='gray')
plt.title("5x5 Sub-region Intensity")
plt.show()
```
